```python
def getKeysList(dict):
    list = []
    for key in dict.keys():
        list.append(key)
          
    return list
```

```python
def getValueList(dict):
    list = []
    for val in dict.values():
        list.append(val)
          
    return list
```

```python
ipl_2020_values = getValueList(sorted_ipl_2020)
print(ipl_2020_values)

ipl_2021_values = getValueList(sorted_ipl_2021)
print(ipl_2021_values)

ipl_2022_values = getValueList(sorted_ipl_2022)
print(ipl_2022_values)
```

```python
data=['265', '12.0 Cr','15.0 Cr','15.0 Cr','15.0 Cr','15.0 Cr','12.50 Cr','12.50 Cr','12.50 Cr','12.50 Cr']
data1 = [['264', '12.0 Cr','15.0 Cr','15.0 Cr','15.0 Cr','15.0 Cr','12.50 Cr','12.50 Cr','12.50 Cr','12.50 Cr']]
dff = pd.DataFrame(data1, columns=['playerId','2022','2021','2020','2019','2018','2017','2016','2015','2014'])

dff = dff.append(pd.DataFrame([data], 
              columns=['playerId','2022','2021', '2020', '2019', '2018', '2017', '2016', '2015', '2014']), 
              ignore_index=True)
dff
```

```python
val = price_df.values.tolist()

```

```python
final_list = []
for row in val:
  newlist = []

  for item in row:
    if type(item) == type(1):
      dig = item
    else:
      l = len(item)
      suff = (item[l - 2:])

      if suff == 'Cr':
        mod_string = item[:l - 3]
        dig = float(mod_string) * 10
      elif suff == ' L':
        mod_string = item[:l - 2]
        dig = float(mod_string) * 0.1
    # else:
    #   dig = item

    newlist.append(dig)

  final_list.append(newlist)
# final_list
```

```python

auction_price_history_df = pd.DataFrame(final_list, columns=['playerId','2022','2021','2020','2019','2018','2017','2016','2015','2014'])
auction_price_history_df
```

```python
id_name = pd.read_csv('https://github.com/Dhanashri30/Thesis/blob/d50cf7cb1bf9fc867a6bf198fe8d5589f49e1337/id_name.csv?raw=true')
id_name.rename(columns  = {'Player Id': 'playerId'}, inplace=True)
id_name.head()
```

```python
auction_price_history = pd.merge(id_name, auction_price_history_df, on='playerId')
auction_price_history.head()
```

```python
auction_price_history[['playerId']] = auction_price_history[['playerId']].astype(int)
auction_price_history.head()
```

```python
auction_price_history.to_csv('auction_price_history_latest.csv',index=False)
```
